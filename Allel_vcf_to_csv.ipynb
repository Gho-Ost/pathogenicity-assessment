# VCF to csv using scikit-allel

```python
import allel
```

## Data read

```python
data = allel.read_vcf("data/EE_sample/EE_sample.vcf")
data
```

## Data to dataframe

```python
df = allel.vcf_to_dataframe("data/EE_sample/EE_sample.vcf", ["CHROM", "POS", "ID", "REF", "ALT", "QUAL", "FILTER", "INFO"])
df
```

## Data to csv

```python
allel.vcf_to_csv("data/EE_sample/EE_sample.vcf", "data/EE_sample/EE_sample_allel.csv", ["CHROM", "POS", "ID", "REF", "ALT", "QUAL", "FILTER", "INFO"])
```
